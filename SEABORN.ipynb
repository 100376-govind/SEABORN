```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.scatterplot(data = penguins , x ="flipper_length_mm" , y= "body_mass_g" , hue = "island")
```

```python
sns.set_style("ticks")
sns.scatterplot(data = penguins , x ="flipper_length_mm" , y= "body_mass_g" , hue = "sex")
sns.despine(left = True) # Remove the spine
```

```python
sns.set_context("talk")  # context = paper , notebook(default) , talk , poster
sns.scatterplot(data = penguins , x ="flipper_length_mm" , y= "body_mass_g" , hue = "sex")
```

```python
sns.set_palette("Set2")  # Palette = Set1 ,Set2 , Set3 , deep , muted , bright , pastel , dark , Colorblind , Qualitative(Set1,Set2,Set3,Pastel1,Dark2,Colorblind,Accent), Sequential(Blues,Greens,Purples,OrRd,YIGnBu,BuPu,Greys) , Diverging(coolwarm,RdBu,BrBG,PiYG,Spectral,terrain_r)
sns.scatterplot(data = penguins , x ="flipper_length_mm" , y= "body_mass_g" , hue = "sex")
```

```python
sns.scatterplot(data = penguins , x ="species" , y= "body_mass_g" , hue = "island" , style ="sex" , alpha = 0.5)
```

```python
# STRIP PLOT :

sns.stripplot(data = penguins , x ="species" , y= "body_mass_g" ,jitter = True , hue = "island", dodge = True)
```

```python
# SWARM PLOT :
sns.set_context("talk")
sns.set_palette("Set1")
sns.swarmplot(data = penguins , x ="species" , y= "body_mass_g" , hue = "island")
sns.despine()
```

```python
# HISTOGRAM :

sns.set_context("notebook")
sns.histplot(data = penguins , x = "body_mass_g", hue  = "sex" , multiple = "stack")
```

```python
# REGRESSION PLOT;
sns.regplot(data = penguins , x ="body_mass_g" , y = "flipper_length_mm")
```

```python
# LINE PLOT :

sns.set_style("whitegrid")
sns.lineplot(data = penguins , x = "body_mass_g" , y = "flipper_length_mm" , hue = "island" , style = "sex")
```

```python
# JOINT PLOT :

sns.jointplot(data = penguins , x = "body_mass_g" , y = "flipper_length_mm")
```

```python
sns.jointplot(data = penguins , x = "body_mass_g" , y = "flipper_length_mm" , hue = "island")
```

```python
sns.jointplot(data = penguins , x = "body_mass_g" , y = "flipper_length_mm" , hue = "island" , kind = "kde")
```

```python
# BAR PLOT :
sns.barplot(data = penguins , y = "body_mass_g" , x = "species" , hue = "sex" , estimator = np.sum)
 
```

```python
# COUNT PLOT:
sns.countplot(data = penguins , x = "species", hue = "island" ,palette = "Set1")
 
```

```python
# BOX PLOT:

sns.boxplot(data = penguins , x = "species" , y = "flipper_length_mm" , hue = "island")
```

```python
# VIOLIN PLOT:

sns.violinplot(data = penguins , x = "species" , y = "flipper_length_mm" , hue = "island")
```

```python
# KDE PLOT:

sns.kdeplot(data = penguins , x = "body_mass_g", hue = "species" , fill = "True")
```

```python
# HEATMAP:

columns = ["bill_length_mm" , "bill_depth_mm" , "flipper_length_mm" , "body_mass_g"]

sns.heatmap(data = penguins[columns].corr(), annot =True, cmap = "Blues" ,vmin =0.2)
```

```python
# RUG PLOT:

sns.rugplot(data = penguins, y = "body_mass_g" , hue ="species" ,palette = "Set1" , height =0.2)
```

```python
# PAIR PLOT:

sns.pairplot(data = penguins, hue ="sex", palette = "Set1", diag_kind = "kde")
```

```python
#PAIR GRID :

graph = sns.PairGrid(data = penguins ,hue = "sex" , palette = "Set1")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
